## <font color=#007ACC>3) Tratamento e Análise dos Dados </font>

```python
# Read the Excel file from the database folder
file_path = 'database/colaboradores.xlsx'
df = pd.read_excel(file_path, sheet_name='CLT')
```

```python
df.head(20)
```

```python
departamentos = df["CLASSIFICAÇÃO"].unique().tolist()
departamentos
```

```python
setor: list = df["SETOR DE TRABALHO"].unique().tolist()
setor
```

```python
numero_funcionarios: int = df.shape[0]
numero_funcionarios
```

```python
formacao_status = df["STATUS"].unique().tolist()
formacao_status
```

```python
# Vamos criar uma coluna onde teremos o status real que será 1 ou zero apenas se a coluna de Formação Atual for igual a Ensino Superior ou Pós Gradução e a coluna de Status for igual a Concluído
df['status_formacao_atual'] = df.apply(
    lambda x: 1 if (x['FORMAÇÃO ATUAL'] in ['ENSINO SUPERIOR', 'PÓS-GRADUAÇÃO']) and (x['STATUS'] == 'CONCLUÍDO') else 0, axis=1)
```

```python
df
```

## <font color=#007ACC>4) Análise dos Dados </font>

---

### <font color=gray> 4.1) Percentual e Qtde de Funcionários com Graduação </font>

```python
# Calcular a porcentagem de funcionários com formação concluída
porcentagem_concluida: float = df['status_formacao_atual'].mean() * 100
print(round(porcentagem_concluida,2))
```

```python
df_formados: pd.DataFrame = df[df['status_formacao_atual'] == 1]
df_formados
```

```python
funcionarios_graduados: int = df_formados.shape[0]
funcionarios_graduados
```

---

### <font color=gray> 4.2) Percentual de Funcionários Graduados com Setor de Trabalho Exceto Instrumentação </font>

```python
df_formados_exceto_instrumentacao = df_formados[df_formados['SETOR DE TRABALHO'] != 'INSTRUMENTAÇÃO']
df_formados_exceto_instrumentacao
```

```python
qtde_func_formados_sem_ser_instrumentacao = df_formados_exceto_instrumentacao.shape[0]
qtde_func_formados_sem_ser_instrumentacao
```

```python
porcentagem_formados_exceto_instrumentacao: float = (qtde_func_formados_sem_ser_instrumentacao / numero_funcionarios) * 100
print(round(porcentagem_formados_exceto_instrumentacao, 2))
```
